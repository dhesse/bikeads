```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('201402_trip_data.csv', parse_dates=[2,5])
```

```python
starts = data.groupby('Start Station').size()
```

```python
# get the index of the start station with most trips recorded
max_index = starts.index[starts == starts.max()][0]
```

```python
print "Most trips start from", max_index
```

```python
# get the data according to the start station with most trips recorded
data_max = pd.DataFrame(data[data['Start Station'] == max_index])
```

# Exploratory customers plot

```python
plt.ylabel("counts")
plt.title("Customers' destinations")
data_max[data_max['Subscription Type'] == 'Customer'].groupby('End Terminal').size().plot(kind='bar')
plt.savefig("customers.pdf")
```

# Exploratory subsrcibers plot

```python
plt.title("Subscribers' destintations")
plt.ylabel("counts")
data_max[data_max['Subscription Type'] == 'Subscriber'].groupby('End Terminal').size().plot(kind='bar')
plt.savefig("subscribers.pdf")
```

# Subscribers by zip

```python
subscriber_data = data_max[data_max['Subscription Type'] == 'Subscriber']
```

```python
subscriber_data['Zip Code'][65]
```

```python
zip_counts = subscriber_data.groupby('Zip Code').size()
zip_counts.sort()
```

```python
by_zip = pd.DataFrame(dict((z, subscriber_data[subscriber_data['Zip Code'] == z].groupby('End Terminal').size())
                           for z in subscriber_data['Zip Code'].unique()
                           if (subscriber_data['Zip Code'] == z).sum() >= zip_counts[-5])).fillna(0)
```

```python
ax = by_zip.plot(kind='bar', stacked='true', title='')
ax.set_ylabel('counts')
ax.set_title("Subscribers' trip destinations, by cyclist's zip code")
plt.savefig("by_zip.pdf")
```
